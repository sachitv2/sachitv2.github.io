Altair Visualization Project

This project uses Altair to create a set of visualizations exploring investment data from IFC (International Finance Corporation) and MIGA (Multilateral Investment Guarantee Agency).

```python
import altair as alt

# Load the data from the URL
data_url = "https://raw.githubusercontent.com/deekshamanoharrao/deekshamanoharrao.github.io/main/_data/IFC_Investment_Services_Projects.csv"

# Define the brush for the selection
brush = alt.selection_interval(encodings=['x', 'y'])

# Create the heatmap chart with the brush selection
heatmap = alt.Chart(data_url).mark_rect().encode(
    x='Industry:N',
    y=alt.Y('Country:N', sort='-x'),  # Sort countries by the x-value (e.g., Industry)
    color='sum(Total IFC investment as approved by Board(Million - USD)):Q'
).properties(
    title='Dashboard for IFC investement dataset',
    width=600,
    height=alt.Step(20)  # Set a custom step for the height of each mark (country)
).add_selection(
    brush
)

# Create the first bar chart that updates based on the selection in the heatmap
bar_chart_1 = alt.Chart(data_url).mark_bar().encode(
    x='Country:N',
    y='sum(Total IFC investment as approved by Board(Million - USD)):Q'
).transform_filter(
    brush
).properties(
    title='Brush selection display by Country',
    width=600,
    height=200
)

# Create the second bar chart that updates based on the selection in the heatmap
bar_chart_2 = alt.Chart(data_url).mark_bar().encode(
    x='Industry:N',
    y='sum(Total IFC investment as approved by Board(Million - USD)):Q'
).transform_filter(
    brush
).properties(
    title='Brush selection display by Industry',
    width=600,
    height=200
)

# Display the charts
combined_charts1_primary = alt.vconcat(heatmap, bar_chart_1, bar_chart_2)

# Show the combined charts
combined_charts1_primary
```

```python
myJekyllDir = '/Users/Deeksha/Downloads/deekshamanoharrao.github.io/assets/json/'

# Save the combined chart to JSON
combined_charts1_primary.save(myJekyllDir + 'Interactive_Visualization.json')
```

```python
import altair as alt

# Load the IFC dataset
ifc_data_url = "https://raw.githubusercontent.com/deekshamanoharrao/deekshamanoharrao.github.io/main/_data/IFC_Investment_Services_Projects.csv"

# Geographical Distribution for IFC and MIGA
ifc_geo_chart = alt.Chart(ifc_data_url).mark_bar().encode(
    x='Country:N',
    y='sum(Total IFC investment as approved by Board(Million - USD)):Q',
    color=alt.value('blue')
).properties(
    title='IFC Investment by Country',
    width=300,
    height=200
)

miga_geo_chart = alt.Chart(miga_data_url).mark_bar().encode(
    x='Host Country:N',
    y='sum(Gross Issued Amount):Q',
    color=alt.value('green')
).properties(
    title='MIGA Investment by Country',
    width=300,
    height=200
)

geo_combined_chart = alt.hconcat(ifc_geo_chart, miga_geo_chart).resolve_scale(y='independent')

# Sectoral Distribution for IFC and MIGA
ifc_sector_chart = alt.Chart(ifc_data_url).mark_bar().encode(
    x='Industry:N',
    y='sum(Total IFC investment as approved by Board(Million - USD)):Q',
    color=alt.value('blue')
).properties(
    title='IFC Investment by Sector',
    width=300,
    height=200
)

miga_sector_chart = alt.Chart(miga_data_url).mark_bar().encode(
    x='Business Sector:N',
    y='sum(Gross Issued Amount):Q',
    color=alt.value('green')
).properties(
    title='MIGA Investment by Sector',
    width=300,
    height=200
)

sector_combined_chart = alt.hconcat(ifc_sector_chart, miga_sector_chart).resolve_scale(y='independent')



miga_time_series_chart = alt.Chart(miga_data_url).mark_line().encode(
    x='Issued FY:O',
    y='sum(Gross Issued Amount):Q',
    color=alt.value('green')
).properties(
    title='MIGA Investment Time Series',
    width=400,
    height=200
)


# Combine all charts
combined_charts2_contextual = alt.vconcat(geo_combined_chart, sector_combined_chart, miga_time_series_chart)

# Show the combined charts
combined_charts2_contextual
```

```python
# Save the combined chart to JSON
combined_charts2_contextual.save(myJekyllDir + 'Contextual_Visulization.json')
```
